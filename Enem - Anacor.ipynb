```python
df = pd.DataFrame(pd.read_csv("/content/drive/MyDrive/MBA - Data Science & Analytics/TCC/Códigos/MICRODADOS_ENEM_2022.csv", sep=';', encoding="ISO-8859-1"))
```

```python
df.head()
```

```python
df_enem = df[['NU_INSCRICAO', 'TP_COR_RACA', 'TP_ESCOLA', 'IN_TREINEIRO', 'TP_DEPENDENCIA_ADM_ESC', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO', 'Q006']]
df_enem.head()
```

```python
df_enem.describe()
```

```python
df_enem=df_enem.dropna(subset=['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO'])
df_enem
```

```python
df_enem['TP_DEPENDENCIA_ADM_ESC'] = df_enem['TP_DEPENDENCIA_ADM_ESC'].fillna(0)
df_enem
```

```python
df_enem['NM_DEPENDENCIA_ADM_ESC'] = np.where(df_enem['TP_DEPENDENCIA_ADM_ESC'] == 1, "Federal",
                             np.where(df_enem['TP_DEPENDENCIA_ADM_ESC'] == 2, "Estadual",
                             np.where(df_enem['TP_DEPENDENCIA_ADM_ESC'] == 3, "Municipal",
                             np.where(df_enem['TP_DEPENDENCIA_ADM_ESC'] == 4, "Privada", "Nao Identificado"))))
df_enem.head()
```

```python
df_enem['TREINEIRO'] = np.where(df_enem['IN_TREINEIRO'] == 1, "Sim",
                             np.where(df_enem['IN_TREINEIRO'] == 0, "Nao", "Nao Identificado"))
df_enem.head()
```

```python
df_enem['NM_COR_RACA'] = np.where(df_enem['TP_COR_RACA'] == 1, "Branca",
                             np.where(df_enem['TP_COR_RACA'] == 2, "Preta",
                             np.where(df_enem['TP_COR_RACA'] == 3, "Parda",
                             np.where(df_enem['TP_COR_RACA'] == 4, "Amarela",
                             np.where(df_enem['TP_COR_RACA'] == 5, "Indigena",
                             np.where(df_enem['TP_COR_RACA'] == 6, "Nao Dispoe da Informacao", "Nao Declarado"))))))
df_enem.head()
```

```python
df_enem['TIPO_ESCOLA_EM'] = np.where(df_enem['TP_ESCOLA'] == 1, "Nao Respondeu",
                            np.where(df_enem['TP_ESCOLA'] == 2, "Publica",
                            np.where(df_enem['TP_ESCOLA'] == 3, "Privada", "Nao Identificado")))
df_enem.head()
```

```python
df_enem['RENDA_FAMILIA'] = np.where(df_enem['Q006'] == 'A', "Nenhuma Renda",
                             np.where(df_enem['Q006'] == 'B', "Ate 1.212,00",
                             np.where(df_enem['Q006'] == 'C', "De 1.212,01 ate 1.818,00",
                             np.where(df_enem['Q006'] == 'D', "De 1.818,01 ate 2.424,00",
                             np.where(df_enem['Q006'] == 'E', "De 2.424,01 ate 3.030,00",
                             np.where(df_enem['Q006'] == 'F', "De 3.030,01 ate 3.636,00",
                             np.where(df_enem['Q006'] == 'G', "De 3.636,01 ate 4.848,00",
                             np.where(df_enem['Q006'] == 'H', "De 4.848,01 ate 6.060,00",
                             np.where(df_enem['Q006'] == 'I', "De 6.060,01 ate 7.272,00",
                             np.where(df_enem['Q006'] == 'J', "De 7.272,01 ate 8.484,00",
                             np.where(df_enem['Q006'] == 'K', "De 8.484,01 ate 9.696,00",
                             np.where(df_enem['Q006'] == 'L', "De 9.696,01 ate 10.908,00",
                             np.where(df_enem['Q006'] == 'M', "De 10.908,01 ate 12.120,00",
                             np.where(df_enem['Q006'] == 'N', "De 12.120,01 ate 14.544,00",
                             np.where(df_enem['Q006'] == 'O', "De 14.544,01 ate 18.180,00",
                             np.where(df_enem['Q006'] == 'P', "De 18.180,01 ate 24.240,00",
                             np.where(df_enem['Q006'] == 'Q', "Acima de 24.240,00", "Nao Declarado")))))))))))))))))
df_enem.head()
```

```python
df_enem['NOTA_FINAL'] = df_enem[["NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_MT", "NU_NOTA_REDACAO"]].mean(axis=1)
df_enem.head()
```

```python
df_enem = df_enem.loc[df_enem['IN_TREINEIRO'] == 0]
df_enem
```

```python
df_enem_anacor = df_enem[['NU_INSCRICAO', 'NM_DEPENDENCIA_ADM_ESC', 'NM_COR_RACA', 'TIPO_ESCOLA_EM', 'RENDA_FAMILIA', 'Q006', 'NOTA_FINAL', 'TREINEIRO']]
df_enem_anacor
```

```python
df_enem_anacor['SUCESSO'] = np.where(df_enem_anacor['NOTA_FINAL'] < 200, 'Nota Ruim',
                            np.where(df_enem_anacor['NOTA_FINAL'] > 690, 'Nota Boa', 'Nota Regular'))
df_enem_anacor
```
