```python
df_geral['SITU_FUNCIONAMENTO'] = np.where(df_geral['TP_SITUACAO_FUNCIONAMENTO'] == 1, "Em Atividade",
                             np.where(df_geral['TP_DEPENDENCIA'] == 2, "Paralisada",
                             np.where(df_geral['TP_DEPENDENCIA'] == 3, "Extinta (ano do Censo)",
                             np.where(df_geral['TP_DEPENDENCIA'] == 4, "Extinta em Anos Anteriores", "Nao Identificado"))))
```

```python
# Read CSV into pandas

dependencia = df1['NM_DEPENDENCIA']
matriculados = df1['TOTAL_MAT']

# Figure Size
fig, ax = plt.subplots(figsize =(7, 3))

# Horizontal Bar Plot
ax.barh(dependencia, matriculados, color='olivedrab')

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 5)

# Show top values
ax.invert_yaxis()

# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             '{:,}'.format(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
# Show Plot
plt.xticks([])
plt.title("Quantidade de Matriculados na Ed. básica por Dependência Administrativa")
plt.savefig('/content/drive/MyDrive/MBA - Data Science & Analytics/TCC/Gráficos - Resultados/graph1.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
df_graph2_ = (df_graph2
      .groupby("NM_DEPENDENCIA")["SITU_FUNCIONAMENTO"]
      .value_counts(normalize=True)
      .mul(100)
      .round(2)
      .unstack())
df_graph2_
```

```python
df_graph2_.fillna(0, inplace=True)
```

```python
df_graph2_
```

```python
fig, ax = plt.subplots(figsize = (5,5))
ax.bar(df_graph2_.index, df_graph2_["Em Atividade"], label = "Em Atividade", width = 0.3)
ax.bar(df_graph2_.index, df_graph2_["Extinta (ano do Censo)"], bottom = df_graph2_.iloc[:, 0], label = "Extinta (ano do Censo)", width = 0.3)
ax.bar(df_graph2_.index, df_graph2_["Extinta em Anos Anteriores"], bottom = df_graph2_.iloc[:, 1], label = "Extinta em Anos Anteriores", width = 0.3)
ax.bar(df_graph2_.index, df_graph2_["Nao Identificado"], bottom = df_graph2_.iloc[:, 2], label = "Não Identificado", width = 0.3)
ax.bar(df_graph2_.index, df_graph2_["Paralisada"], bottom = df_graph2_.iloc[:, 3], label = "Paralisada", width = 0.3)
```

```python
ax.containers
```

```python
# Accessing the heights from each rectangle
for c in ax.containers:
    for v in c:
        print(v.get_height().round(2))
```

```python
# Add labels
for c in ax.containers:
    labels = [str(round(v.get_height(), 2)) + "%" if v.get_height() > 0 else '' for v in c]
    ax.bar_label(c,
                 label_type='center',
                 labels = labels,
                 size = 14) # add a container object "c" as first argument
# Remove spines
for s in ["top", "right"]:
    ax.spines[s].set_visible(False)
# Add labels
ax.tick_params(labelsize = 12, labelrotation = 0)
ax.set_ylabel("%", size = 14)
ax.set_xlabel("Dependência", size = 14)
# Add legend
ax.legend(labels = ["Em Atividade", "Extinta (ano do Censo)", "Extinta em Anos Anteriores", "Não Identificado", "Paralisada"],
          fontsize = 12,
          title = "Situação Funcionamento",
          title_fontsize = 14,
          bbox_to_anchor = [1.1, 0.7])
# Fix legend position
# ax.legend_.set_bbox_to_anchor([0.55, 0.7])
ax.get_figure().savefig('/content/drive/MyDrive/MBA - Data Science & Analytics/TCC/Gráficos - Resultados/graph2.png', dpi=300, bbox_inches='tight')
fig
```
